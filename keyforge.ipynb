```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load deck info:
deck_df = pd.read_csv("decks.csv", index_col=0)
deck_df
```

```python
# Set colors:
set_colors = {
    "CotA": "firebrick",
    "AoA": "mediumblue",
    "WC": "darkviolet",
    "MM": "black",
    "DT": "deepskyblue",
    "WoE": "olivedrab",
    "GR": "red",
}
```

```python
# House colors:
house_colors = {
    "brobnar": "darkorange",
    "dis": "magenta",
    "logos": "deepskyblue",
    "mars": "yellowgreen",
    "sanctum": "mediumblue",
    "shadows": "grey",
    "untamed": "darkgreen",
    "star alliance": "gold",
    "saurian": "aquamarine",
    "unfathomable": "blueviolet",
    "ekwidon": "red",
}
```

```python
# Plot number of decks per house/stat as heatmap:
set_house_df = pd.DataFrame(0, index=set_colors.keys(), columns=house_colors.keys())
for idx, row in deck_df.iterrows():
    for house in row[2:5]:
        set_house_df.loc[row[1], house] += 1

plt.figure(figsize = (12,2))
_ = sns.heatmap(set_house_df, cmap="RdYlGn", annot=True, cbar=False)
```

```python
# Load match info:
match_df = pd.read_csv("matches.csv")
match_df
```

```python
# Create match heatmap:
match_hm = pd.DataFrame(index=deck_df.index, columns=deck_df.index)
for _, row in match_df.iterrows():
    match_hm.loc[row[1], row[2]] = row[3]
    match_hm.loc[row[2], row[1]] = row[4]
match_hm.fillna(-1, inplace=True)

plt.figure(figsize = (15,15))
_ = sns.heatmap(match_hm, cmap="RdYlGn", square=True, annot=True, cbar=False)
```

## Deck stats

```python
# Create stats per deck:
for idx, row in deck_df.iterrows():
    deck_df.loc[idx, "plays"] = sum(match_hm.loc[idx,:] > -1)
    deck_df.loc[idx, "wins"] = sum(match_hm.loc[idx,:] == 3)
    deck_df.loc[idx, "keys_forged"] = sum(match_hm.loc[idx, match_hm.loc[idx,:] > -1])
    deck_df.loc[idx, "key_diff"] = deck_df.loc[idx, "keys_forged"] - sum(match_hm.loc[match_hm.loc[:,idx] > -1, idx])
    deck_df.loc[idx, "win_rate"] = round(deck_df.loc[idx, "wins"]/deck_df.loc[idx, "plays"]*100, 1)
    deck_df.loc[idx, "avg_keys_forged"] = round(deck_df.loc[idx, "keys_forged"]/deck_df.loc[idx, "plays"], 1)
    deck_df.loc[idx, "avg_key_diff"] = round(deck_df.loc[idx, "key_diff"]/deck_df.loc[idx, "plays"], 1)
```

```python
# Plot stats as heatmap:
plt.figure(figsize = (10,8))
deck_labels = [f"{idx}: {row[0]} ({row[1]} - {row[2]} / {row[3]} / {row[4]})" for idx, row in deck_df.iterrows()]
_ = sns.heatmap(deck_df.iloc[:,[6, 10, 11, 12]], yticklabels=deck_labels , cmap="RdYlGn", annot=True, fmt="g", cbar=False)
```

```python
_ , ax = plt.subplots(1, figsize=(10,6))

# Plot for each deck win% vs SAS (color by set):
for deck_set in set_colors:
    deck_sas = deck_df.loc[deck_df["set"] == deck_set, "SAS"]
    deck_win_rate = deck_df.loc[deck_df["set"] == deck_set, "win_rate"]
    set_data = ax.scatter(deck_sas, deck_win_rate, c=set_colors[deck_set], label = deck_set, zorder = 10)
    set_data.set_clip_on(False)

# Plot trendline + R2
x = list(deck_df.loc[deck_df["win_rate"].notnull(), "SAS"])
y = list(deck_df.loc[deck_df["win_rate"].notnull(), "win_rate"])
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax.plot(x, p(x), color="grey", linewidth=0.5)

plt.xlabel("SAS")
plt.ylabel("Win rate [%]")
ax.set_ylim(ymin=0, ymax=100)
plt.legend()
plt.show()
```

## Set stats

```python
# Create stats per set:
set_df = pd.DataFrame(index=set_colors.keys())
for idx in set_df.index:
    set_df.loc[idx, "decks"] = sum(deck_df["set"] == idx)
    for column in ["plays", "wins", "keys_forged", "key_diff"]:
        set_df.loc[idx, column] = sum(deck_df.loc[deck_df["set"] == idx, column])
    set_df.loc[idx, "win_rate"] = round(set_df.loc[idx, "wins"]/set_df.loc[idx, "plays"]*100, 1)
    set_df.loc[idx, "avg_keys_forged"] = round(set_df.loc[idx, "keys_forged"]/set_df.loc[idx, "plays"], 1)
    set_df.loc[idx, "avg_key_diff"] = round(set_df.loc[idx, "key_diff"]/set_df.loc[idx, "plays"], 1)
    set_df.loc[idx, "avg_deck_sas"] = round(np.mean(deck_df.loc[deck_df["set"] == idx, "SAS"]), 1)
    set_df.loc[idx, "avg_deck_win_rate"] = round(np.mean(deck_df.loc[deck_df["set"] == idx, "win_rate"]), 1)
    set_df.loc[idx, "std_deck_win_rate"] = round(np.std(deck_df.loc[deck_df["set"] == idx, "win_rate"]), 1)
```

```python
# Set heatmaps: stats - match plays - match win rates
fig , (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))

# Set match heatmaps: 
_ = sns.heatmap(set_df.iloc[:,[0, 1, 5, 6, 7, 8]], cmap="RdYlGn", ax=ax1, square=True, annot=True, fmt="g", cbar=False)

# Create set match heatmaps: Number of plays and number of wins
set_plays_hm = pd.DataFrame(0, index=set_df.index, columns=set_df.index)
set_wins_hm = pd.DataFrame(0, index=set_df.index, columns=set_df.index)
for set_1 in set_df.index:
    for set_2 in set_df.index:
        decks_1 = deck_df.index[deck_df["set"] == set_1]
        decks_2 = deck_df.index[deck_df["set"] == set_2]
        plays = (match_hm.loc[decks_1, decks_2] > -1).sum().sum()
        wins = (match_hm.loc[decks_1, decks_2] == 3).sum().sum()
        win_rate = round(wins/plays*100) if plays > 0 else -1
        set_plays_hm.loc[set_1, set_2] = plays
        set_wins_hm.loc[set_1, set_2] = win_rate
_ = sns.heatmap(set_plays_hm, cmap="RdYlGn", ax=ax2, square=True, annot=True, fmt="g", cbar=False)
_ = sns.heatmap(set_wins_hm, cmap="RdYlGn", ax=ax3, square=True, annot=True, fmt="g", cbar=False)
```

```python
# Set plots: Average win rates vs SAS - Changes in SAS overtime - Changes in win rate overtime
fig , ax = plt.subplots(1, 3, figsize=(20,6))

# Average win rates vs SAS (color by set):
for deck_set in set_colors:
    x = set_df.loc[deck_set, "avg_deck_sas"]
    y = set_df.loc[deck_set, "avg_deck_win_rate"]
    yerr = set_df.loc[deck_set, "std_deck_win_rate"]
    ax[0].scatter(x, y, c=set_colors[deck_set], label = deck_set)
    ax[0].errorbar(x, y, yerr=yerr, fmt="-o", color=set_colors[deck_set])
ax[0].set_xlabel("SAS")
ax[0].set_ylabel("Win rate [%]")
ax[0].set_xlim(xmin=50, xmax=80)
ax[0].set_ylim(ymin=0, ymax=100)
ax[0].legend()

# Changes in SAS overtime:
ax[1].plot(set_df.index, set_df["avg_deck_sas"])
ax[1].set_ylabel("SAS")
ax[1].set_ylim(ymin=50, ymax=80)

# Changes in win rate overtime:
ax[2].plot(set_df.index, set_df["win_rate"])
ax[2].set_ylabel("Win rate [%]")
_ = ax[2].set_ylim(ymin=20, ymax=80)
```

```python
# Find set pair that has been played the least:
def find_least_played_set_pair(set_plays_hm):
    for set1, row in (set_plays_hm == set_plays_hm.min().min()).iterrows():
        for set2, val in row.items():
            if val:
                return (set1, set2)

# Rank sets by usage:
def rank_set_decks_by_plays(deck_df, set_id):
    set_series = deck_df.loc[deck_df["set"] == set_id, "plays"]
    set_series.sort_values(ascending=True, inplace=True)
    return set_series

# Compute next game to play:
def find_next_game(deck_df, set_plays_hm, match_hm):
    (set1, set2) = find_least_played_set_pair(set_plays_hm)
    set1_ranked = rank_set_decks_by_plays(deck_df, set1)
    set2_ranked = rank_set_decks_by_plays(deck_df, set2)
    for deck1, _ in set1_ranked.items():
        for deck2, _ in set2_ranked.items():
            if match_hm.loc[deck1, deck2] == -1:
                return [deck1, deck2]
    
deck_pair = find_next_game(deck_df, set_plays_hm, match_hm)
deck_df.loc[deck_pair,]
```

## House stats

```python
# Create stats per house:
house_df = pd.DataFrame(index=house_colors.keys())
for idx in house_df.index:
    decks_w_house = sum([deck_df[f"house{i}"] == idx for i in [1,2,3]]) == 1
    house_df.loc[idx, "decks"] = sum(decks_w_house)
    for column in ["plays", "wins", "keys_forged", "key_diff"]:
        house_df.loc[idx, column] = sum(deck_df.loc[decks_w_house, column])
    house_df.loc[idx, "win_rate"] = round(house_df.loc[idx, "wins"]/house_df.loc[idx, "plays"]*100, 1)
    house_df.loc[idx, "avg_keys_forged"] = round(house_df.loc[idx, "keys_forged"]/house_df.loc[idx, "plays"], 1)
    house_df.loc[idx, "avg_key_diff"] = round(house_df.loc[idx, "key_diff"]/house_df.loc[idx, "plays"], 1)
    house_df.loc[idx, "avg_deck_sas"] = round(np.mean(deck_df.loc[decks_w_house, "SAS"]), 1)
    house_df.loc[idx, "avg_deck_win_rate"] = round(np.mean(deck_df.loc[decks_w_house, "win_rate"]), 1)
    house_df.loc[idx, "std_deck_win_rate"] = round(np.std(deck_df.loc[decks_w_house, "win_rate"]), 1)
```

```python
# House heatmaps: stast - match plays - match win rates
fig , (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(22,6))

# Plot stats as heatmap:
_ = sns.heatmap(house_df.iloc[:,[0, 1, 5, 6, 7, 8]], cmap="RdYlGn", ax=ax1, square=True, annot=True, fmt="g", cbar=False)

# Create house match heatmaps: Number of plays and number of wins
house_plays_hm = pd.DataFrame(0, index=house_df.index, columns=house_df.index)
house_wins_hm = pd.DataFrame(0, index=house_df.index, columns=house_df.index)
for house_1 in house_df.index:
    for house_2 in house_df.index:
        decks_1 = deck_df.index[sum([deck_df[f"house{i}"] == house_1 for i in [1,2,3]]) == 1]
        decks_2 = deck_df.index[sum([deck_df[f"house{i}"] == house_2 for i in [1,2,3]]) == 1]
        plays = (match_hm.loc[decks_1, decks_2] > -1).sum().sum()
        wins = (match_hm.loc[decks_1, decks_2] == 3).sum().sum()
        win_rate = round(wins/plays*100) if plays > 0 else -1
        house_plays_hm.loc[house_1, house_2] = plays
        house_wins_hm.loc[house_1, house_2] = win_rate
_ = sns.heatmap(house_plays_hm, cmap="RdYlGn", ax=ax2, square=True, annot=True, fmt="g", cbar=False)
_ = sns.heatmap(house_wins_hm, cmap="RdYlGn", ax=ax3, square=True, annot=True, fmt="g", cbar=False)
```

```python
# House plot: Average win rates vs SAS
fig , ax = plt.subplots(1, 1, figsize=(7,6))
for deck_house in house_colors:
    x = house_df.loc[deck_house, "avg_deck_sas"]
    y = house_df.loc[deck_house, "avg_deck_win_rate"]
    yerr = house_df.loc[deck_house, "std_deck_win_rate"]
    ax.scatter(x, y, c=house_colors[deck_house], label = deck_house)
    ax.errorbar(x, y, yerr=yerr, fmt="-o", color=house_colors[deck_house])
ax.set_xlabel("SAS")
ax.set_ylabel("Win rate [%]")
ax.set_xlim(xmin=50, xmax=75)
ax.set_ylim(ymin=0, ymax=100)
ax.legend()
```
